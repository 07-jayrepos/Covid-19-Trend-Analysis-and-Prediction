```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
data = pd.read_csv('Cov-Data-India.csv', sep=',')
data = data[['days' , 'total_cases']]
print(data)
```

```python
x = np.array(data['days']).reshape(-1,1)
y = np.array(data['total_cases']).reshape(-1,1)
```

```python
plt.plot(y, '-m')
plt.show()
```

```python
polFeat = PolynomialFeatures(degree=7)
val = polFeat.fit_transform(x)
val
```

```python
pd.DataFrame(val)
```

```python
model = linear_model.LinearRegression()
model.fit(val , y)
```

```python
Yp = model.predict(val)
Yp
```

```python
model.intercept_
```

```python
plt.scatter(x , y , label="Actual")
plt.plot(x , Yp , 'r' , label="Predicted")
plt.xlabel("Number of Days")
plt.ylabel("Cases")
plt.legend(loc="upper left")
plt.show()
```

```python
model.score(val , y)
```
